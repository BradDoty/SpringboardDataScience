**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

```python
import locale

locale.setlocale(locale.LC_ALL, 'en_GB')

def averagePriceChange(data, begin, end, region=None):
    ''' Calculate and return change in average house prices from begin time period to end time period for region(s).
        begin and end can be scalar dates or a range of dates for an annual or other time period average.
        region can be a single district name, a list of districts, or None to calculate increases for all districts.
        Returns a percent change for a single district, or a list of change for all the districts requested.
    '''
    price_change = {}
    
    if region is None:
        region = data.columns
    elif type(region) is not list:
        region = [region]
        
    if type(begin) is None or type(end) is None or type(data) is None or len(data) == 0:
        print ("data (a Pandas dataframe) and begin and end (dates or lists of dates) are all required arguments.")
        return None
    if type(begin) is not list:
        begin = [begin]
    if type(end) is not list:
        end = [end]
        
    #print ("region=",region,"begin=",begin, "end=", end)
            
    for district in region:
        first_price = data[district][(data.index >= begin[0]) & (data.index <= begin[-1])].agg(np.mean)
        last_price = data[district][(data.index >= end[0]) & (data.index <= end[-1])].agg(np.mean)
        #print (first_price, last_price)
        price_change[district] = round(100 * ((last_price / first_price) - 1), 1)
    
    #print("Price Change: (in %)\n", price_change)        
    return price_change
            
## Test the function            
#averagePriceChange(props, begin="1995-01-01", end="2021-04-01")
#averagePriceChange(props, "1995-01-01", "2021-04-01", region='Croydon')
#averagePriceChange(props, begin=["1995-01-01"], end=["2021-04-01"], region='Croydon')
#averagePriceChange(props, "2020-01-01", ["2021-04-01"], region=['Islington','Croydon', 'Kensington & Chelsea'])
#averagePriceChange(props, ["2000-01-01", "2000-12-01"], ["2020-01-01", "2020-12-01"], region=['City of London','Croydon'])


## Let's answer the original question first.
## Which boroughs have seen the greatest increase in average housing prices averaged over 12 months in the last 2 decades.
## We'll take the top 5 in the time period 2000 - 2020.

price_change = averagePriceChange(props, begin=["2000-01-01", "2000-12-01"], end=["2020-01-01", "2020-12-01"])
change = pd.Series(data=price_change)
print("Highest percent gainers, 2000-2020\n")
print(change.nlargest(5, keep='all'))
print("The most expensive districts, like Chelsea and City of London are not in this list.")

## Let's also look at the bottom 5 from 2000 - 2020
print("\nLowest percent gainers, 2000-2020\n")
print(change.nsmallest(5, keep='all'))

## And let's look how the districts have fared during the pandemic, March 2020 - April 2021 (latest available)
pandemic_change = averagePriceChange(props, begin='2020-03-01', end='2021-04-01')
pandemic = pd.Series(data=pandemic_change)
print("\nPercentage gainers 3/2020 to 4/2021")
print(pandemic.nlargest(6, keep='all'))

print("\nPercentage losers 3/2020 to 4/2021")
print(pandemic.nsmallest(6, keep='all'))

print("Some of London's wealthiest districts have sold off houses during the pandemic.")
```
